```python
year2012africa = africa[africa['year'] == 2012]
year2012africa
```

```python
#create a scatter plot of life expectancy versus fertility for the African continent in 2012
plt.figure(figsize=(10,10))
sns.scatterplot(x='life_expectancy', y='fertility', data=year2012africa);
```

```python
#Remake the plot from the previous exercises but this time use color to distinguish the different regions of Africa to see if this explains the clusters. 
plt.figure(figsize=(10,10))
sns.scatterplot(x='life_expectancy', y='fertility', data=year2012africa, hue='region');
```

```python
#Create a table showing the country and region for the African countries (use select) that in 2012 had fertility rates of 3 or less and life expectancies of at least 70.
table = year2012africa[(year2012africa['fertility']<3) & (year2012africa['life_expectancy']>70)]
table
```

```python
#create a dollars_per_day variable, which is defined as gdp/population/365.
africa
```

```python
africa['dollars_per_day'] = (africa['gdp'])/(africa['population'])/365
```

```python
africa
```

```python
africa.dropna(inplace=True)
```

```python
africa
```

```python
#Create the dollars_per_day variable for African countries for the year 2010.
daydollars2010 = africa[africa['year'] == 2010]
```

```python
daydollars2010
```

```python
#Create a smooth density plot of dollars per day from daydollars
plt.figure(figsize=(10,10))
sns.distplot(a=daydollars2010.dollars_per_day, hist=False);
```

```python
#Create the dollars_per_day variable for African countries for the year 2010.
daydollars1970 = africa[africa['year'] == 1970]
```

```python
daydollars1970
```

```python
#Create a smooth density plot of dollars per day from daydollars
plt.figure(figsize=(10,10))
sns.distplot(a=daydollars1970.dollars_per_day, hist=False);
```

```python
#Create a smooth density plot of dollars per day for 1970 and 2010
plt.figure(figsize=(10,10))
sns.distplot(a=daydollars1970.dollars_per_day, hist=False, label='1970')
sns.distplot(a=daydollars2010.dollars_per_day, hist=False, label='2010');
```

```python
#Make a scatter plot of infant_mortality versus dollars_per_day for countries in the African continent.
plt.figure(figsize=(20,20))
sns.scatterplot(x='infant_mortality',y='dollars_per_day',data=daydollars2010, hue='region',legend='full');
sns.scatterplot(x='infant_mortality',y='dollars_per_day',data=daydollars2010, hue='country',legend='full');
```
